```python
import pandas as pd
import numpy as np
import datetime as dt

%load_ext autoreload
%autoreload 2

pd.options.display.max_rows = 2000
pd.options.display.max_columns = 2000
```

```python
# slicing by expiration month
# pass in a number of cols of a product

FC = pd.read_csv("../data/FC1.csv")
exp_months = FC["Exp_Month"].unique()

for m in range(2,8):
    m_read = pd.read_csv(f"../data/FC{m}.csv")
    FC = FC.append(m_read, ignore_index=True)

#print(FC)

# store the new dataframes in a dictionary but can easily export all of these to csvs. Would just be a lot
gf_months = {}

for month in exp_months:
    mon_df = FC.loc[FC["Exp_Month"] == month]
    mon_df.set_index("Date", inplace=True)
    gf_months[month] = mon_df
```

```python
# slicing by expiration month
# pass in a number of cols of a product

LC = pd.read_csv("../data/LC1.csv")
exp_months = LC["Exp_Month"].unique()

for m in range(2,8):
    m_read = pd.read_csv(f"../data/LC{m}.csv")
    LC = LC.append(m_read, ignore_index=True)

#print(LC)

# store the new dataframes in a dictionary but can easily export all of these to csvs. Would just be a lot
lc_months = {}

for month in exp_months:
    mon_df = LC.loc[LC["Exp_Month"] == month]
    mon_df.set_index("Date", inplace=True)
    lc_months[month] = mon_df
```

```python
gf_aug = [gf_months[month] for month in exp_months if 'AUG' in month]
lc_feb = [lc_months[month] for month in exp_months if 'FEB' in month]
```

```python
gf = pd.concat(gf_aug)
lc = pd.concat(lc_feb)
```

```python
gf.tail()
```

```python
df = gf.merge(lc, how='inner', left_index=True, right_index=True, suffixes=["_FC", "_LC"])
```

```python
zc = pd.read_csv("../data/C.csv").set_index('Date')
zc.index = pd.to_datetime(zc.index)
```

```python
df = df.merge(zc, how='inner', left_index=True, right_index=True, suffixes=["", "_ZC"])
```

```python
df
```

```python
import matplotlib.pyplot as plt
import datetime

years = {}
for i in range(21):
    yr_df = df[(pd.to_datetime(df.index) > datetime.datetime(2000 + i, 5, 1)) & (pd.to_datetime(df.index) < datetime.datetime(2000 + i, 9, 1))]
    yr_df['Feeder Profit/CWT'] = (6 * 400 * yr_df['Settle_LC'].values - 2 * 50 * yr_df['Settle'].values - 3 * 500 * yr_df['Settle_FC'].values) / 1500
    years[2000 + i] = yr_df
```

```python
for year, yr_df in years.items():
    print(f"YEAR: {year}")
    print(f"EXPIRATION: {yr_df.index[-1]}")
    plt.figure(figsize=(15, 5))
    plt.plot(yr_df['Feeder Profit/CWT'])
    plt.show()
```
